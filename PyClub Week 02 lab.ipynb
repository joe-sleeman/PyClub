```python
"""PyClub week 02 lab"""
import csv
from itertools import groupby
import matplotlib.pyplot as plt

# set stuff up
col_xp_lvl = []
col_set_up = []
col_difficulty = []
col_had_fun = []
colors = ['limegreen', 'green', 'darkgreen', 'yellowgreen']

# read file, put in list of columns
with open('data.csv', 'rb') as file:
    reader = csv.reader(file, delimiter=",")
    for row in reader:
        col_xp_lvl.append(row[0])
        col_set_up.append(row[1])
        col_difficulty.append(row[2])
        col_had_fun.append(row[3])

# list of lists
columns = [col_xp_lvl, col_set_up, col_difficulty, col_had_fun]

# loop through columns
for col in columns:
    # pop the title at index 0
    title = col.pop(0)
    # sort
    col.sort()
    # get unique values out of column
    labels = list(set(col))
    labels.sort()
    sizes = ([len(list(group)) for key, group in groupby(col)])
    plt.title(title)
    plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()
```
